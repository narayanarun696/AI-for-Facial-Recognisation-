```python
wrinkle=load_model("/content/drive/MyDrive/CSV/VGG16_Wrinkle(1).h5")

frickle=load_model("/content/drive/MyDrive/CSV/Freakles(1).h5")

glass=load_model("/content/drive/MyDrive/CSV/VGG16_Glasses(1).h5")

hair_color=load_model("/content/drive/MyDrive/CSV/VGG16_HairColor(1).h5")

hair_top=load_model("/content/drive/MyDrive/CSV/VGG16_Hair_Top(1).h5")
```

```python

import matplotlib.pyplot as plt
import cv2,glob
items = glob.glob('/content/drive/MyDrive/TestImages/*')
#items.sort(key=natural_sort_key)
print(len(items))
img_lst = []
for i in range(len(items)):
  image = cv2.imread(items[i])
  image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  image = cv2.resize(image, (96,96),interpolation=cv2.INTER_AREA)
  img_lst.append(image)
```

# **Testing the model with random images**

# **Wrinkles**

```python
label_to_text={0:'no wrinkes',1:'wrinkles'}
from IPython.display import clear_output
import time
for k in range(len(img_lst)):
  predicted_class=wrinkle.predict(tf.expand_dims(img_lst[k],0)).argmax()
  print(f'predicted label have {label_to_text[predicted_class]}')
  plt.imshow(img_lst[k])
  plt.show()
  time.sleep(1)
  clear_output(wait=True)
```

```python
predicted_class=[]
for k in range(len(img_lst)):
  predicted_class.append(wrinkle.predict(tf.expand_dims(img_lst[k],0)).argmax())
def unique(list1):
 
    # intilize a null list
    unique_list = []
     
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print (x)
unique(predicted_class)
```
